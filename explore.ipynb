```python
import pandas as pd
import numpy as np
```

### Set first unnamed column as index column

```python
data = pd.read_csv('dataset.csv', index_col='Unnamed: 0')
data = data[~data.track_id.duplicated(keep="first")] # get rid of duplicated songs
# round up tempo
data['tempo'] = data['tempo'].apply(np.ceil)
data.head()
```

```python
data.describe()
```

From above, we can see we have 114,000 entries(songs)

```python
data.columns, len(data.columns)
```

```python
genres = data['track_genre'].unique()
genres, len(genres)
```

```python
# find songs with similar tempos/bpm
def same_tempo(tempo):
    return data.loc[data['tempo'] == tempo]

# find songs with songs within a tempo range of 5
# eg for a tempo of 120, return all songs with a 
# tempo of 115-125
def same_tempo_range(tempo):
    tempos = list(range(tempo-5, tempo+6))
    return data.loc[data['tempo'].isin(tempos)]

same_tempo(125)
same_tempo_range(125)
```

```python
genres
```

```python
# find songs in certain genre
# find songs at certain bpm, and certain genre
def get_songs_in_genre(genre):
    return data.loc[data.track_genre == genre]

# find songs at certain bpm, and certain genre
def same_tempo_with_genre(tempo, genre):
    return same_tempo(tempo).loc[data.track_genre == genre]


def get_songs_around_tempo_in_genre(tempo, genre):
    return same_tempo_range(tempo).loc[data.track_genre == genre]

same_tempo_with_genre(130, 'hip-hop')
get_songs_around_tempo_in_genre(130, 'hip-hop')
```

## explore music popularity

```python
data.popularity
```

```python
# define popular score as >= 50 popularity
POPULAR = 75
```

```python
# try to get popular songs
data.loc[data.popularity >= POPULAR]
```

```python
# get popular songs in a certain genre
def get_popular_songs_in_genre(genre):
    return get_songs_in_genre(genre).loc[data.popularity >= POPULAR]

get_popular_songs_in_genre('hip-hop')
```

```python
# get popular songs around a certain tempo
def get_popular_songs_around_tempo(tempo):
    return same_tempo_range(tempo).loc[data.popularity >= POPULAR]

get_popular_songs_around_tempo(125)
```

```python
def get_popular_songs_around_tempo_in_genre(tempo, genre):
    return get_popular_songs_around_tempo(tempo).loc[data.track_genre == genre]

get_popular_songs_around_tempo_in_genre(120, 'hip-hop')
```
